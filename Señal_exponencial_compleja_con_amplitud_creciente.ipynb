```python
#Seal exponencial compleja con amplitud creciente
%matplotlib inline
#para graficar en el entorno
import sympy as sym #importo librera sympy
sym.init_printing() #inicializo para mostrar en pantalla
t, sigma, omega= sym.symbols("t sigma omega",real=True) #defino las variables simblicas
s= sigma+1j*omega #sobre cargo operador para definir parte imaginaria
x= sym.exp(s*t) #funcin exponencial simblica desde sympy
x

#Ahora, vamos a establecer valores numricos a omega y sigam para una amplitud creciente
y=x.subs({omega:4, sigma:0.4})
print(r"x(t)= ",y)

#Ahora, vamos a graficar la parte real e imaginaria vs el tiempo
sym.plot(sym.re(y),(t,0,2*sym.pi),ylabel=r"Re{$e^{st}$}",ylim=[-20,15])
sym.plot(sym.im(y),(t,0,2*sym.pi),ylabel=r"Im{$e^{st}$}",ylim=[-20,15])
```
