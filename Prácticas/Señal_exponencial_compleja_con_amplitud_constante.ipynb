```python
#Seal exponencial compleja con amplitud constante
%matplotlib inline
#para graficar en el entorno
import sympy as sym #importando libreria sympy
sym.init_printing() #inicializando para mostrar en pantalla
t, sigma, omega = sym.symbols("t sigma omega",real=True) #defino mis variables simblicas
s= sigma+1j*omega #sobrecargo operador para definir la parte imaginaria
x=sym.exp(s*t) #funcin exponencial simbolica desde sympy
x

#Vamos a darle valores a sigma y omega par que la amplitud sea constante
y=x.subs({omega:9, sigma:0})
print(r"x(t)= ",y)

#Ahora vamos a plotear la seal, tanto su parte real como su parte imagniria vs el tiempo
sym.plot(sym.re(y),(t,0,2*sym.pi),ylabel= r"Re{$e^{st}$}",ylim=[-1,1])
sym.plot(sym.im(y),(t,0,2*sym.pi),ylabel=r"Im{$e^{st}$}", ylim=[-1,1])
```
